```python
import requests
```

```python
url="https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json"
response = requests.get(url)
data = response.json()
```

```python
print(type(data))
```

```python
print(data.keys())
```

```python
objects=data['objects']
print(type(objects))
print(len(objects))
```

```python
techniques=[]
for obj in objects:
    if obj.get('type')=="attack-pattern":
        techniques.append(obj)
```

```python
print("Total Techniques:",len(techniques))
print(techniques[0])
```

```python
def threat_score(tech):
    score = 5
    
    name = tech.get("name", "").lower()
    
    if "credential" in name:
        score += 3
    if "execution" in name:
        score += 2
    if "privilege" in name:
        score += 3
    if "persistence" in name:
        score += 2
    if "lateral" in name:
        score += 2

    description = tech.get("description", "").lower()
    
    if "administrator" in description:
        score += 2
    if "remote" in description:
        score += 2
    if "bypass" in description:
        score += 2
    if "stealth" in description:
        score += 1

    return score
```

```python
scored_techniques = []
for tech in techniques:
    name=tech.get("name","Unknown")
    score=threat_score(tech)
    scored_techniques.append((name,score))
```

```python
scored_techniques.sort(key=lambda x: x[1], reverse=True)
print("Top 10 High-Risk Techniques:\n")

for name, score in scored_techniques[:10]:
    print(f"{name} — Score: {score}")
```

```python
print("\nCritical Threats (Score >= 8.9):\n")

for name, score in scored_techniques:
    if score >= 8.9:
        print(f"{name} — Score: {score}")
```

### How your scoring helps analysts?

#### This scoring system automates threat prioritization by assigning higher scores to techniques associated with credential theft, privilege escalation, remote attacks, and security bypass behaviors. By ranking techniques based on risk indicators in both names and descriptions, analysts can focus on high-impact threats first, improving response efficiency and reducing investigation time.
